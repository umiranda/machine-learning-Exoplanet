```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC 
```

```python
df = pd.read_csv("/Users/ulissesmiranda/Documents/GitHub/machine-learning-challenge/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(30)
```

```python
X = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_period_err1', 'koi_period_err2',
       'koi_time0bk', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact',
       'koi_impact_err1', 'koi_impact_err2', 'koi_duration',
       'koi_duration_err1', 'koi_duration_err2', 'koi_depth', 'koi_depth_err1',
       'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2',
       'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2',
       'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_steff_err1',
       'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2',
       'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec',
       'koi_kepmag']]
y = df["koi_disposition"]
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=["koi_disposition"])
y = df["koi_disposition"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_train.head()
```

```python
# Scaling data
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)


X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y_encoder = LabelEncoder().fit(y_train)
```

```python
y_train_encoded = y_encoder.transform(y_train)
y_test_encoded = y_encoder.transform(y_test)
y_train_encoded
```

```python
print(y_train)
```

```python
for x in y_train_encoded:
    print(x)
```

```python
model = SVC(kernel='linear')
model.fit(X_train_scaled,y_train_encoded)
predictions = model.predict(X_test_scaled)
model
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train_encoded)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test_encoded)}")
```

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test_encoded, predictions,
                           target_names =["confirmed", "false positive", "candidate"]))
```

```python
# Create the GridSearchCV model
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'gamma':['scale', 'auto'],
              'kernel': ['linear']}  
   
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
```

```python
# fitting the model for grid search 
grid.fit(X_train_scaled, y_train_encoded) 
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
